```python
import pandas as pd

df = pd.read_csv('MT745584.txt', header=None)
amino_acids = ('A', 'C', 'G', 'T')

stack = [''] # BFS로 접근
over_minsup = {'string': [], 'sup': []}
minsup = 10

display(df)
```

```python
try_count = 0

while(len(stack) != 0):
    try_count += 1
    node = stack.pop()

    if node != '':
        sup = df.apply(lambda x: node in x[0], axis='columns')
        sup = sup.sum()
    else:
        sup = len(df)

    if sup < minsup:
        continue
    else:
        over_minsup['string'].append(node)
        over_minsup['sup'].append(sup)
        for acid in amino_acids: stack.append(f'{node}{acid}')

    print(f'{try_count:7d} strings have been checked. {len(stack):3d} strings are in stack.', end='\r')
```

```python
result = pd.DataFrame(over_minsup)
display(result)
```

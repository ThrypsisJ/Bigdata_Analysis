```python
import pandas as pd
import numpy as np
from random import randint
import pickle

# 인공 데이터셋 생성
# 각 포인트는 10차원의 벡터로 구성
# label=0인 포인트는 [0~115,  0~115,  85~200, 0~115,  85~200, 85~200, 0~115,  85~200, 0~115,  0~115 ]
# label=1인 포인트는 [85~200, 85~200, 0~115,  85~200, 0~115,  0~115,  85~200, 0~115,  85~200, 85~200]
# label=0인 포인트와 label=1인 포인트 사이에는 각 차원의 값에 대하여 30 정도의 회색지대가 존재함

points_a = [np.array([randint(0,115), randint(0,115), randint(85,200), randint(0,115), randint(85,200), randint(85,200), randint(0,115), randint(85,200), randint(0,115), randint(0,115)]) for _ in range(5000)]
points_b = [np.array([randint(85,200), randint(85,200), randint(0,115), randint(85,200), randint(0,115), randint(0,115), randint(85,200), randint(0,115), randint(85,200), randint(85,200)]) for _ in range(5000, 10000)]
points = points_a + points_b

labels = [0 if idx < 5000 else 1 for idx in range(10000)]

dataset = pd.DataFrame({ 'data': points, 'labels': labels })
dataset = dataset.sample(frac=1).reset_index(drop=True)
display(dataset)

with open('./dataset.pkl', 'wb') as file: pickle.dump(dataset, file)
```

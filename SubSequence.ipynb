```python
import pandas as pd

df = pd.read_csv('MT745584.txt', header=None)
amino_acids = ('A', 'C', 'G', 'T')

idxs_base = {}
for data in df.iloc[:,0].to_list():
    idxs_base[data] = 0

stack = [('', idxs_base)] # DFS로 접근
over_minsup = {'sequence': [], 'sup': []}
minsup = 498

display(df)
```

```python
def check(target, sub:tuple, idx_dic:dict):
    max_length = len(target[0])
    try:
        index = sub[1][target[0]]
        suffix = sub[0][-1]
        for idx in range(index, max_length):
            if target[0][idx] == suffix:
                idx_dic[target[0]] = idx+1
                return True
    except: pass
    return False

try_count = 0

while(len(stack) != 0):
    try_count += 1

    node = stack.pop()
    idxs = {}

    if node[0] != '':
        sup = df.apply(check, axis='columns', sub=node, idx_dic=idxs)
        sup = sup.sum()
    else:
        idxs = idxs_base
        sup = len(df)

    if sup < minsup:
        continue
    else:
        over_minsup['sequence'].append(node[0])
        over_minsup['sup'].append(sup)
        for acid in amino_acids: stack.append((f'{node[0]}{acid}', idxs))
    
    print(f'{try_count:7d} sequences have been checked. {len(stack):3d} sequences are in stack.', end='\r')
print(f'{try_count:7d} sequences have been checked. {len(stack):3d} sequences are in stack.')
```

```python
result = pd.DataFrame(over_minsup)
display(result)
```
